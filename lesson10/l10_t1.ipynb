```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
warnings.filterwarnings("ignore")  # Игнорируем предупреждения
```

```python
titanic = pd.read_csv(r"C:\Users\София\DataProject\lesson9\Titanic.csv")
```

```python
titanic.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis = 1, inplace = True) # Удаление ненужных столбцов
```

```python
mean_age = titanic['Age'].mean()
titanic['Age'].fillna(mean_age, inplace = True) # Заполнение пропущенных значений в столбце возраста
```

```python
titanic
```

```python
sns.histplot(data = titanic, x = 'Age', bins = 20, color = 'purple')  # График количества пассажиров по возрастам
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.title('Количество пассажиров по возрасту')
```

```python
sns.countplot(data = titanic, x = 'Pclass', color = 'purple') # График количества пассажиров по классу каюты
plt.xlabel('Класс каюты')
plt.ylabel('Количество')
plt.title('Количество пассажиров по классу каюты')
```

```python
sns.scatterplot(data=titanic, x='Pclass', y='Fare', color = 'purple')  # График зависимости стоимости билета от класса каюты
plt.xlabel('Класс каюты')
plt.ylabel('Стоимость')
plt.title('Зависимость стоимости билета от класса каюты')
```

```python
color = '#8c4db6'
sns.scatterplot(data = titanic, x = 'Pclass', y = 'Age', hue = 'Survived', palette = [color])  # График отношения возраста и класса каюты
plt.xlabel('Класс каюты')
plt.ylabel('Возраст')
plt.title('Отношение возраста и класса каюты')
```

```python
colors = ['#6a0dad', '#b26fd0']
sns.countplot(data = titanic, x = 'Survived', hue = 'Sex', palette = colors) # График выживших и не выживших мужчин и женщин
plt.xlabel('Не выжившие/Выжившие')
plt.ylabel('Количество')
plt.title('Выжившие и не выжившие мужчины и женщины')
```

```python
class_counts = titanic['Pclass'].value_counts() # Создание нового датафрейма с количеством пассажиров по классу каюты
colors = ['#6a0dad', '#8c4db6', '#b26fd0'] 
plt.pie(class_counts, labels = ['3 класс', '1 класс', '2 класс'], autopct = '%1.1f%%', colors = colors) # Построение круговой диаграммы количества пассажиров по классу каюты
plt.title('Количество пассажиров по классу каюты')
```
